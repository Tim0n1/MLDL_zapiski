## Basic Concepts

### Figure

The top level container for all the plot elements. A Figure Object contains the whole plot.
You can have more than one Figure object.
Reference: [Figure Class](https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure)

<img src="images/empty_figure.png" width="300">

### Axes (and Subplots)
The Axes represent an **individual drawing plot**, with its own coordinate system.

**A figure can contain many Axes, but a given Axes object can only be placed in one Figure.**

Subplots are Axes instances with additional methods to facilitate generating and manipulating a set of Axes within a figure. Every Subplot instance is a valid Axes Object.

Reference: 
	
- [Axes Class](https://matplotlib.org/api/axes_api.html#matplotlib.axes.Axes)

- [Figure.add_subplot](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_subplot)

- [Figure.add_axes](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_axes)

- [What are the differences between add_axes and add_subplot?](https://stackoverflow.com/questions/43326680/what-are-the-differences-between-add-axes-and-add-subplot)


````python
fig = plt.figure()
fig.set_facecolor("#B15A5A")

# add a new axes at the position 1 on a grid of 2 rows and 1 column.
fig.add_subplot(211)

# add a new axes at the position 2 on a grid of 2 rows and 1 column.
fig.add_subplot(212)

````

Create figure and axes with [matplotlib.pyplot.subplots](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)

This wrapper method makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call. This is the preferred way of creatting fugure and axes objects.

````python
fig, axes = plt.subplots(nrows=2)
fig.set_facecolor("#B15A5A")

plt.show()
print(f'Number of created axes: {len(axes)}')

````

### Axes title and labels

Each Axes can have a title ([Axes.set_title](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html#matplotlib.axes.Axes.set_title)) and labels ([Axes.set_xlabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html), [Axes.set_ylabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html#matplotlib.axes.Axes.set_ylabel)) 

````python
fig, axes = plt.subplots()
fig.set_facecolor("#B15A5A")

axes.set_title("y=sqrt(x)")

axes.set_xlabel("X",fontsize=14, color="yellow")
axes.set_ylabel("Y", rotation=0,fontsize=14, color="yellow")

axes.xaxis.set_label_coords(1.05, 0.02)
axes.yaxis.set_label_coords(0, 1.05)
````

#### Controlling style of text and labels using a dictionary
Reference: 
- [Text properties and layout](https://matplotlib.org/stable/tutorials/text/text_props.html)
- [fontdict](https://matplotlib.org/stable/gallery/text_labels_and_annotations/text_fontdict.html#sphx-glr-gallery-text-labels-and-annotations-text-fontdict-py)

````python
fig, axes = plt.subplots()
fig.set_facecolor("#B15A5A")

axes.set_title("y=sqrt(x)", 
	fontdict={
		'fontsize': 30,
		'fontweight': 'bold',
		'color': 'red',
		'backgroundColor':'#999',
		'verticalalignment': 'bottom',
		# 'horizontalalignment': 'right'
	 },
	 loc='right'
)
# note that 'horizontalalignment' aligns text relative to its center position by default


axes.set_xlabel("X",fontsize=14, color="yellow")
axes.set_ylabel("Y", rotation=0,fontsize=14, color="yellow")

axes.xaxis.set_label_coords(1.05, 0.02)
axes.yaxis.set_label_coords(0, 1.05)
````

### Ticks

Each axis has a ticks. With Matplotlib you can configure all aspects (location, formating, etc.) of the ticks

````python
fig, axes = plt.subplots(nrows=1)
fig.set_facecolor("#B15A5A")

axes.get_xticklabels()[3].set_color("#F3FF00")
````

### Lets put it all together

````python
x = np.array([1,2,3,4])
y1 = x**2
y2 = x**3

axes_names = ['Axes 1', 'Axes 2']

fig, axes = plt.subplots(nrows=2)
fig.set_facecolor("#CCCCCC")

for ax, y, name,color in zip(axes, [y1, y2], axes_names, ["#00FF80","#0080FF"]):
    ax.plot(x, y, color='black')
    ax.set_title(name)
    ax.set_facecolor("#BA5E5E")
    ax.set_ylabel('Y Axis')
    ax.set_xlabel('X Axis')    


plt.setp(axes[0].spines.values(), color="#F3FF00")
plt.setp(axes[1].spines.values(), color="#00F3FF")

axes[0].get_xticklabels()[2].set_color("#F3FF00")
axes[1].get_xticklabels()[2].set_color("#00F3FF")
plt.tight_layout()
plt.show()
````

### Multiple plots in single axes

````python
n = 256
X = np.linspace(-np.pi,np.pi,n)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color=(0,1,0,1)) # equiv to: color='green', alpha=1.00


plt.show()
````
